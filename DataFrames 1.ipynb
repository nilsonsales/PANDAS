### Drop Rows with Null Values

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.tail(3)
```

```python
# Changing 'how' to 'all', removes only in ALL values are NA
nba.dropna(how="all", inplace=True)
```

```python
# Remove COLUMNS with any NA value
nba.dropna(axis=1)
```

```python
# Remove NA rows only considering Salary column
nba.dropna(subset=["Salary"])
```

### Fill in Null Values with the .fillna() Method

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.fillna(0).head()
```

```python
# Doing it only for a column
nba["Salary"].fillna(0, inplace=True)
nba.head()
```

```python
nba["College"].fillna("No College", inplace=True)
nba.head()
```

### The .astype() Method

```python
nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"].fillna(0, inplace=True)
nba["College"].fillna("None", inplace=True)
nba.head(6)
```

```python
nba.dtypes

# Another way
nba.info()
```

```python
nba["Age"] = nba["Age"].astype("int")
nba["Number"] = nba["Number"].astype("int")
nba.head(3)
```

```python
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
nba["Team"] = nba["Team"].astype("category")

nba.info()
```

### Sort a DataFrame with the .sort_values() Method, Part I

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values("Age", ascending=False)

nba.sort_values("Salary", ascending=False)

nba.sort_values("Name", ascending=True, inplace=True)
nba.head(3)
```
